```python
from pyspark.sql import SparkSession

# Example: 8 cores, 128GB total memory
spark = SparkSession.builder \
    .config("spark.driver.memory", "2g") \
    .config("spark.executor.memory", "18g") \
    .config("spark.executor.instances", 7) \
    .getOrCreate()
```

```python
path = "datasets/coco2017/annotations/captions_train2017.json"
captions_train = spark.read.json(path)
```

```python
captions_train.printSchema()
captions_train.select("annotations").show(1, truncate=True)
```
